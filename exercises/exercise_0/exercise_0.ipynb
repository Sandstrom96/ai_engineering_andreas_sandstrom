### b) Draw a line chart of quantity for each year. Is there a year that should be skipped?

```python
df[df["Year"] == 2017]
```

```python
df_quantity = df.groupby("Year")["Quantity"].sum().reset_index()
# drop 2017 dÃ¥ den endast har en mÃ¥nads data
df_quantity = df_quantity[df_quantity["Year"] != 2017]
df_quantity.plot(x="Year", y="Quantity")
```

### c) Draw a line chart of average CO2 emissions for same years that as in b)

```python
df_emissions = df.groupby("Year")["Avg_CO2"].mean().reset_index()
df_emissions = df_emissions[df_emissions["Year"] != 2017]
df_emissions.plot(x="Year", y="Avg_CO2")
```

### d) Draw a line chart of all years and months for import

```python
df["Date"] = pd.to_datetime(df[["Year", "Month"]].assign(day=1))
ax = df.plot(x="Date", y="Import", legend=False)
ax.set(title="New cars imported per year/month", xlabel="Year/month", ylabel="Total imported cars");
```

### e) Draw a line chart of all years and months for average CO2 emissions

```python
ax = df.plot(x="Date", y="Avg_CO2", legend=False)
ax.set(title="Average CO2 emissions per year/month", xlabel="Year/month", ylabel="Total avg CO2");
```

###  f) Draw a line chart of all years and months for electric cars import where it's relevant.

```python
df[["Year", "Import_Electric"]]
```

```python
df_electric_import = df[["Date", "Import_Electric"]].dropna()
ax = df_electric_import.plot(x="Date", y="Import_Electric", legend=False)
ax.set(title="Imported electric cars per year/month", xlabel="Year/month", ylabel="Total imported cars");
```

### g) Draw a line chart of average diesel share per year

```python
df_avg_diesel = df.groupby("Year")["Diesel_Share"].mean().reset_index()
ax = df_avg_diesel.plot(x="Year", y="Diesel_Share", legend=False)
ax.set(title="Average diesel share per year", xlabel="Year", ylabel="Total diesel share");
```
